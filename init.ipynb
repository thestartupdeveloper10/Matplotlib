```python
#dealing with csv data
with open('./data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
#  Counters are a type of dictionary subclass specifically designed to count hashable objects.
    language_counter = Counter()
    for row in csv_reader:
    # splitting the values in the 'LanguagesWorkedWith' column of the current row using ';' as the delimiter
    # and updating the counts of each language.
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
print(language_counter.most_common(15))

language=[]
popularity=[]

for item in language_counter.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
# print(language)
# print(popularity)

# This is done to ensure that the bars will be plotted in descending order of popularity.
language.reverse()
popularity.reverse()

#horizontal bar chart
plt.title('Programming Languages Popularity')
plt.ylabel('Languages')
plt.xlabel('Popularity(users)')

plt.barh(language,popularity)

plt.legend()
plt.tight_layout()

plt.show()
```

stack plots

```python
# regression example
# A stack plot, also known as a stacked area plot, is a type of graph that displays multiple
# datasets stacked on top of each other. It's particularly useful for visualizing the composition
# of a whole while showing how individual parts contribute to the total over time or across categories.


# first,second... minutes
minutes=[1,2,3,4,5,6,7,8,9]

# points scored per minute
player1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

labels=['player1', 'player2', 'player3']

plt.stackplot(minutes,player1,player2,player3,labels=labels)
plt.title('Player Stats Per Minute')
plt.xlabel('Minutes')
plt.ylabel('Points')
plt.legend(loc=(0.07,0.05))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


# Months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

# Usage data for different programming languages
python_usage = [20, 25, 30, 35, 40]
javascript_usage = [15, 20, 22, 25, 30]
java_usage = [10, 12, 15, 18, 20]

# Stack plot
plt.stackplot(months, python_usage, javascript_usage, java_usage, labels=['Python', 'JavaScript', 'Java'])

# Adding labels and legend
plt.title('Programming Language Usage Over Months')
plt.xlabel('Months')
plt.ylabel('Usage (%)')
plt.legend(loc='upper left')

plt.show()
```

#filling areas on line plots

```python
data = pd.read_csv('./fill_data.csv')
data.head()
```

```python

ages=data['Age']
all_dev_salaries = data['All_Devs']
py_dev_salaries = data['Python']
js_dev_salaries = data['JavaScript']

plt.style.use('_classic_test_patch')

plt.ylabel('Salary')
plt.xlabel('Ages')

overal_median= 57287

plt.plot(ages,all_dev_salaries,label='All Devs',linestyle='--')
plt.plot(ages,py_dev_salaries,label='Python')
plt.fill_between(ages,py_dev_salaries,overal_median,
                 where=(py_dev_salaries> overal_median),
                 interpolate=True,color='green',
                 alpha=0.25,label='Above Median')
plt.fill_between(ages,py_dev_salaries,overal_median,
                 where=(py_dev_salaries<= overal_median),
                 color='red',
                 interpolate=True,
                 alpha=0.25,label='Below Median')

plt.fill_between(ages,py_dev_salaries,all_dev_salaries,
                 where=(py_dev_salaries> all_dev_salaries),
                 interpolate=True,
                 alpha=0.25)
plt.fill_between(ages,py_dev_salaries,all_dev_salaries,
                 where=(py_dev_salaries<= all_dev_salaries),
                 interpolate=True,
                  color='yellow',
                 alpha=0.25)


plt.legend()
plt.title('Developers Salary')
plt.tight_layout()
plt.show()
```

Time Series Plots

```python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
# linestyle='solid'= gives it a line
plt.plot_date(dates, y, linestyle='solid')
# format date in a more nice way
plt.gcf().autofmt_xdate()
# change date format
date_format = mpl_dates.DateFormatter('%b,%d %Y')
# setting it
plt.gca().xaxis.set_major_formatter(date_format)
plt.tight_layout()
plt.show()
```
